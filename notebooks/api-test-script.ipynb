```python
if not delete_id:
  print("No delete_id, function is skipped.")
else:
  route = server_url + f"/embedding/delete/{delete_id}"

  print(f"Testing {route}")
  response = requests.delete(route, headers=headers)
  if response.status_code != 200:
    print("Error at: ", response.json())
  else:
    print("Success: ", response.json())
```

```python
# create to delete
route = server_url + "/embedding/insert"
data = [
  {
    "process_id": "666",
    "gdpr_id": "667",
    "topic": "just a test dataset",
    "type": "string",
    "len": 1,
    "document": "there is no dokument in a test dataset."
  }
]
response = requests.put(route, headers=headers, json=data)

if response.status_code != 200:
  print("Error at: ", response.json())
else:
  print("Success: ", response.json())
  delete_id = response.json()[0]
  route = server_url + f"/embedding/delete-many"
  data = {
    "ids": [
      delete_id
    ],
    # "where": {}
  }

  print(f"Testing {route}")
  response = requests.post(route, headers=headers, json=data)
  if response.status_code != 200:
    print("Error at: ", response.json())
  else:
    print("Success: ", response.json())
```
